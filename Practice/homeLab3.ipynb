```python
numeric_cols=df.select_dtypes(include=['int64','float64']).columns
cat_cols=df.select_dtypes(include=['object']).columns

for col in numeric_cols:
    df[col].fillna(df[col].median(),inplace=True)
for col in cat_cols:
    df[col].fillna(df[col].mode()[0],inplace=True)
```

```python
#zscore using the formula method, and printing the outliers
columns=['MSSubClass','LotArea']
mean=df[columns].mean()
std=df[columns].std()

zscores=(df[columns]-mean)/std
outliers=df[zscores.abs().all(axis=1)>3]
outliers[columns]
```

```python
df.shape
```

```python
#zscore using the import and removing the outliers
from scipy.stats import zscore
import numpy as np
columns=['MSSubClass','LotArea']
z_scores=np.abs(zscore(df[columns]))
df_without_outliers=df[(z_scores<3).all(axis=1)]
```

```python
df_without_outliers.shape
```

```python
columns=['MSSubClass','LotArea']

q1=df[columns].quantile(0.25)
q3=df[columns].quantile(0.75)
iqr=q3-q1

lowerbound=1.5-q1*iqr
upperbound=1.5+q3*iqr

df_without_outliers_iqr=df[((df[columns]>=lowerbound)&(df[columns]<=upperbound)).all(axis=1)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr=df.corr(numeric_only=True)['SalePrice'].sort_values(ascending=False)

# top_corr=corr.head(11)
top_corr=corr.index

plt.figure(figsize=(6,4))
sns.heatmap(df[top_corr].corr(),annot=True,cmap='coolwarm')
plt.title('Heatmap of correlation of features with the target variable SalePrice')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

numeric_columns=df.select_dtypes(include=['int64','float64']).columns

# X=df.drop('SalePrice',axis=1)
X=df[numeric_columns]
X=X.drop('SalePrice',axis=1)
y=df['SalePrice']

scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)

pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)

plt.figure(figsize=(6,5))
plt.scatter(X_pca[:,0],X_pca[:,1],c=y,cmap='viridis',s=80)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('PCA in 2 Components')
plt.colorbar(label='SalePrice')
plt.show()
```
