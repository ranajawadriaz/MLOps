```python
class person:
    def __init__(self,firstname,lastname,age):
        self.__firstname=firstname
        self.__lastname=lastname
        self._age=age
    def setAge(self,age):
        self._age=age
    def setFirstName(self,firstname):
        self.__firstname=firstname
    def getFirstName(self):
        return self.__firstname
    def setLastName(self,lastname):
        self.__lastname=lastname
    def getLastName(self):
        return self.__lastname

class faculty(person):
    def __init__(self,courseCount,ext,firstname,lastname,age):
        super().__init__(firstname,lastname,age)
        self.__courseCount=courseCount
        self.__ext=ext
        
    def getAge(self):
        return self._age
    def getCourseCount(self):
        return self.__courseCount
    def setCourseCount(self,count):
        self.__courseCount=count
        
        
    def getExtension(self):
        return self.__ext
        
        
    def setEntension(self,ext):
        self.__ext=ext
        
    
    

class student(person):
    def __init__(self,cgpa,rollnum,firstname,lastname,age):
        super().__init__(firstname,lastname,age)
        
        self.__cgpa=cgpa
        self.__rollnum=rollnum
    def setCGPA(self,cgpa):
        self.__cgpa=cgpa
    def getCGPA(self):
        return self.__cgpa
    def setRollnum(self,rollnum):
        self.__rollnum=rollnum
    def getRollnum(self):
        return self.__rollnum





class undergraduate(student):
    def __init__(self,fypName,supervisorName,cgpa,rollnum,firstname,lastname,age):
        super().__init__(cgpa,rollnum,firstname,lastname,age)
        self.__fypname=fypName
        self.__supervisorName=supervisorName
    def getAge(self):
        return self._age
    def setCGPA(self,cgpa):
        super().setCGPA(cgpa)
        
    def getCGPA(self):
        return super().getCGPA()
    def getFYPname(self):
        return self.__fypname
    def setFYPname(self,name):
        self.__fypname=name
    def setRollNum(self,rollnum):
        super().setRollnum(rollnum)
    def getRollNum(self):
        return super().getRollnum()
    def setSuperVisorName(self,name):
        self.__supervisorName=name
        
    def getSuperVisorName(self):
        return self.__supervisorName
        
    

class graduate(student):
    def __init__(self,thesisTopic,supervisorName,cgpa,rollnum,firstname,lastname,age):
        super().__init__(cgpa,rollnum,firstname,lastname,age)
        self.__thesisTopic=thesisTopic
        self.__supervisorName=supervisorName
    def getAge(self):
        return self._age
    def setCGPA(self,cgpa):
        super().setCGPA(cgpa)
        
    def getCGPA(self):
        return super().getCGPA()
    def getThesisname(self):
        return self.__thesisTopic
    def setThesisname(self,name):
        self.__thesisTopic=name
    def setRollNum(self,rollnum):
        super().setRollnum(rollnum)
        
    def getRollNum(self):
        return super().getRollnum()
    def setSuperVisorName(self,name):
        self.__supervisorName=name
    def getSuperVisorName(self):
        return self.__supervisorName


if __name__=='__main__':
    fac=faculty(3,271,'usama','hassan',30)
    print(fac.getFirstName())
    print(fac.getAge())
    print(fac.getLastName())
    under=undergraduate('AI teacher','kashif',3.5,6874,'rana','jawad',20)
    print(under.getCGPA())
    grad=graduate('AI car','nosheen',3.5,6800,'haseeb','zahid',23)
    
    
        
        
    
        
    
        
        
```
